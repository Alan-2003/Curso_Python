## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def numeroMayor(a,b):
    """ Funcion para obtener el numero mayor
    ## Parametros 
    a = primer numero
    b = segundo numero
    ## return
    regresa el numero mayor o si son iguales
    """
    if a>b :
        return(f'{a} es mayor que {b}')
    elif a<b:
        return(f'{b} es mayor que {a}')
    else:
        return f'{a} es igual a {b}'
     
````

````python
x = int(input("numero1"))
y = int(input("numero2"))

numeroMayor(x, y)
````

## Ejercicio 3:

Un piloto de carreras ha registrado su velocidad cada segundo en la siguiente lista:

velocidad = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]

Hacer una función que imprima en que segundo el piloto rebaso los 210 km/h


**Ejemplo de impresión**

En el segundo `9` el piloto llego a los `223` km/h




````python
def limite_velocidad(velocidades):
   
    segundos = 0
   
    for velocidad in velocidades:
        if velocidad > 210:
            print(f'En el segundo {segundos} el piloto llego a los {velocidad}')
        segundos = segundos + 1
````

````python
x = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]

limite_velocidad(x)
````

## Extra points

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
def numeros_pares_impares(una_lista):
    contador_par = 0
    contador_impar = 0

    for numero in una_lista:
        if numero%2 == 0:
            contador_par += 1
        else:
            contador_impar += 1
    return f'Tiene {contador_par} pares y {contador_impar} impares'
````

````python
numeros=[2, 3, 6, 8, 9]
numeros_pares_impares(numeros)
````
